```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('titan_c.csv')
```

```python
data.head()
```

```python
data['fare'].plot(figsize=(12,8))
data['age'].plot(figsize=(12,8))
plt.show()
```

```python
data.describe()
```

```python
###
```

```python
mean_age = data['age'].mean()
```

```python
mean_fare = data['fare'].mean()
```

```python
std_age = data['age'].std()
```

```python
std_fare = data['fare'].std()
```

```python
data['age_zscore'] = round((data['age']-mean_age)/std_age,2) 
```

```python
data['fare_zscore'] = round((data['fare']-mean_fare)/std_fare,2)
```

```python
data[['age_zscore','fare_zscore']]
```

```python
round(data.describe(),2)
```

```python
data['fare_zscore'].plot(figsize=(12,8))
data['age_zscore'].plot(figsize=(12,8))
plt.show()
```

```python
data.to_csv('stadard_titan.csv',index=False)
```

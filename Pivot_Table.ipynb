```python
import pandas as pd
```

```python
data2 = pd.read_csv('files/table2.csv')
```

```python
data2.head(10)
```

```python
data2.pivot_table(index='Country',columns='Medal',values='Count',aggfunc='sum',fill_value=0,margins=True,margins_name='Total')
```

```python
####
```

```python
data2_group = data2.groupby(['Country','Medal'])
```

```python
agg = data2_group['Count'].sum().unstack(fill_value=0)
```

```python
agg
```

```python
agg['All'] = agg.sum(axis=1)
```

```python
agg.loc['All'] = agg.sum(axis=0)
```

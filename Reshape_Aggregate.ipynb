```python
import pandas as pd
```

```python
data = pd.read_csv('files/titanic.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
pd.crosstab(data['sex'],data['pclass'],margins=True)
```

```python
data_group = data.groupby(['sex','pclass']) #groupby also can do that
```

```python
data_group['pclass'].count().unstack()
```

```python
data_norm = pd.crosstab(data['sex'],data['pclass'],margins=True,normalize=True)
```

```python
data_norm
```

```python
##########################################################
```

```python
pd.crosstab(index=data['sex'],columns=data['pclass'],values=data['age'],aggfunc='mean')
```

```python
###pivot table
```

```python
data.pivot_table(index='sex',columns='pclass',values='age',aggfunc='mean')
```

```python
#groupby method
```

```python
data.groupby(['sex','pclass']).age.mean().unstack()
```

```python
pd.crosstab(index=[data2['Year'],data2['Country']],
           columns= data2['Medal'],aggfunc='count',
           values= data2['Athlete'],margins=True).fillna(0)
```

```python
###using pivot table
```

```python
data2.pivot_table(index=['Year','Country'],columns='Medal',
                 aggfunc='count',values='Athlete',fill_value=0,
                 margins=True)
```

```python
###using groupby
```

```python
data2.groupby(['Year','Country','Medal']).Medal.count().unstack(fill_value=0)
```
